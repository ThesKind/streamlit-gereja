```python
# Tentukan Library Yang Digunakan
import numpy as np 
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# Load Dataset
gaji_dataset = pd.read_csv("data-gaji.csv")
```

```python
gaji_dataset.head(5)
```

```python
gaji_dataset.shape
```

```python
gaji_dataset['Status Penerima Bantuan'].value_counts()
```

```python
# Memisahkan Data Dan Label
X = gaji_dataset.drop(columns='Status Penerima Bantuan', axis=1)
Y = gaji_dataset['Status Penerima Bantuan']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Standarisasi Data
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarized_data = scaler.transform(X)
```

```python
print(standarized_data)
```

```python
X = standarized_data
Y = gaji_dataset['Status Penerima Bantuan']
```

```python
print(X)
print(Y)
```

```python
# Memisahkan Data Training Dan Data Testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Membuat Data Latih Menggunakan Algoritma SVM
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
# Membuat Evaluasi Unutuk Mengukur Tingkat Akurasi 
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Akurasi Data Training Adalah = ', training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Akurasi Data Testing Adalah = ', test_data_accuracy)
```

```python
# Membuat Model Prediksi
input_data = (500000, 300000, 2, 0)

input_data_as_numpy_array = np.array(input_data)

input_data_reshape = input_data_as_numpy_array.reshape(1, -1)

std_data = scaler.transform(input_data_reshape)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if prediction[0] == 0:
    print('Tidak Menerima')
else:
    print('Terima Bantuan')
```
